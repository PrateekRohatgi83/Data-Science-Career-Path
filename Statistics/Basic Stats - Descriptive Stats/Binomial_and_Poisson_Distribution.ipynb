**Binomial Distribution**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
n = 7
```

```python
p = 0.6
```

```python
k = np.arange(0, 8)
k
```

```python
binomial = stats.binom.pmf(k, n, p)
binomial
```

```python
plt.plot(k,binomial,'o-')
plt.title('Binomial: n=%i , p=%.2f' % (n,p), fontsize=10)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Successes')
plt.show()
```

```python
n=len(binomial)
n
```

```python
sum(binomial)
```

**Poisson Distribution**

```python
rate = 6
```

```python
n = np.arange(0,20)
```

```python
poisson = stats.poisson.pmf(n,rate)
```

```python
poisson
```

```python
poisson[3]-poisson[2]
```

```python
1 - (poisson[0]+poisson[1]+poisson[2]+poisson[3])
```

```python
plt.plot(n,poisson,'o-')
plt.title('Poisson: $\lambda$ = %i ' % rate)
plt.xlabel('Number of Customers Arriving in a minute')
plt.ylabel('Probability of Number of Customers arriving in a minute')
plt.show()
```
